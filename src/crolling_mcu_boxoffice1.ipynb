```python
import pandas as pd

# 파일을 열고 잘못된 문자를 무시하여 읽기
with open('/home/jieun/ws/eda/mcu_box_office.csv', 'r', encoding='utf-8', errors='ignore') as file:
    mcu_box_office = pd.read_csv(file, header=0)

mcu_box_office.head()
```

```python
num_li = [6, 9, 10, 16, 18, 20]
for val in num_li:
    db_data_temp["Original Title"][val] = mcu_box_office["Movie"][val]
```

```python
temp_li = []
s="Release Date"
for i, row in enumerate(db_data_temp["Original Title"]):
    # print(row["Original Title"], row["Original Title"] in mcu_box_office["Movie"])
    # print(row in mcu_box_office["Movie"])
    for j, mcu_row in enumerate(mcu_box_office["Movie"]):
        if row.lower() == mcu_row.lower():
            # print(i, j, row, data["Release Date"][j])
            temp_li.append(mcu_box_office[s][j])

print(len(temp_li))

db_data_temp["Release Date"] = temp_li
db_data_temp.head()
```
