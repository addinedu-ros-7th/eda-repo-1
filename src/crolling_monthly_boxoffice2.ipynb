```python
import pandas as pd
dfs = {}
# '해당월' 포맷팅 및 월 변경 함수
def adjust_month(x):
    year = str(x)[:4]
    month = str(x)[4:]
    
    # 상반기 월을 변환
    if month == '6':
        return f"{year}01"  # 06월 -> 01월
    elif month == '5':
        return f"{year}02"  # 05월 -> 02월
    elif month == '4':
        return f"{year}03"  # 04월 -> 03월
    elif month == '3':
        return f"{year}04"  # 03월 -> 02월
    elif month == '2':
        return f"{year}05"  # 02월 -> 05월
    elif month == '1':
        return f"{year}06"  # 01월 -> 06월
    
    # 하반기 월을 변환
    elif month == '12':
        return f"{year}07"  # 12월 -> 07월
    elif month == '11':
        return f"{year}08"  # 11월 -> 08월
    elif month == '10':
        return f"{year}09"  # 10월 -> 09월
    elif month == '9':
        return f"{year}10"  # 09월 -> 10월
    elif month == '8':
        return f"{year}11"  # 08월 -> 11월
    elif month == '7':
        return f"{year}12"  # 07월 -> 12월


# 각 연도와 달에 맞는 DataFrame 생성

for year in range(2008, 2021):  # 2008부터 2020까지
    for half in range(1, 3):  # 1: 상반기 (1-6월), 2: 하반기 (7-12월)
        # 연도별 데이터의 반기에 맞는 테이블 가져오기
        if half == 1:
            data_half = data2008_1 if year == 2008 else globals()[f'data{year}_1']
        else:
            data_half = data2008_2 if year == 2008 else globals()[f'data{year}_2']
        
        # 월에 맞게 테이블 선택 (1부터 6까지는 하향식 선택)
        for month_index in range(6, 0, -1):  # 6부터 1까지
            df_name = f"df_{year}_{month_index + (6 if half == 2 else 0)}"
            dfs[df_name] = data_half[month_index][[
                "순위", "영화명", "개봉일", "매출액", "매출액 점유율", "누적매출액", 
                "관객수", "누적관객수", "스크린수", "상영횟수", "대표국적", "국적", "배급사"
            ]]
            # 해당월 추가
            dfs[df_name].insert(0, "해당월", int(f"{year}{month_index + (6 if half == 2 else 0)}"))
            
            # "순위"가 "합계"인 행을 제거
            dfs[df_name] = dfs[df_name][dfs[df_name]['순위'] != '합계']

# 모든 데이터프레임을 하나로 합치기
combined_df = pd.concat(dfs.values(), ignore_index=True)

# 월 변경 함수 적용
combined_df['해당월'] = combined_df['해당월'].apply(adjust_month)

combined_df

# # 결합된 데이터 출력
# combined_df

```
